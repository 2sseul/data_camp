```python
import pandas as pd

pd.options.display.float_format = '{:,.2f}'.format

df_apart = pd.read_csv('제주_아파트_매매_실거래가.csv', encoding='utf-8').reset_index()
df_apart = df_apart.drop(['level_0', 'index', 'NO'], axis=1)
df_office = pd.read_csv('제주_오피스텔_매매_실거래가.csv', encoding='utf-8').reset_index()
df_office = df_office.drop(['level_0', 'index', 'NO'], axis=1)

df_office

# 거래금액 숫자로 변환
# 쉼표 제거하고 정수형으로 변환
df_apart['거래금액'] = df_apart['거래금액(만원)'].str.replace(",", "").astype(int)*10000
df_office['거래금액'] = df_office['거래금액(만원)'].str.replace(",", "").astype(int)*10000

# 평(㎡ → 평)
df_apart['전용면적(평)'] = df_apart['전용면적(㎡)'] * 0.3025
df_office['전용면적(평)'] = df_office['전용면적(㎡)'] * 0.3025

# 평당가격 계산
df_apart['평당가격'] = df_apart['거래금액'] / df_apart['전용면적(평)']
df_office['평당가격'] = df_office['거래금액'] / df_office['전용면적(평)']

# 계약년도/계약월 분리
df_apart['계약년도'] = df_apart['계약년월'].astype(str).str[:4]
df_apart['계약월'] = df_apart['계약년월'].astype(str).str[4:].str.zfill(2)
df_office['계약년도'] = df_office['계약년월'].astype(str).str[:4]
df_office['계약월'] = df_office['계약년월'].astype(str).str[4:].str.zfill(2)

display(df_apart)
display(df_office)
```

```python
df_apart.to_csv('아파트.csv')
df_office.to_csv('오피스텔.csv')
```

```python
# 필터링 테스트
filtered_df = df_apart[
    (df_apart['거래금액'] <= 300000000) &
    (df_apart['전용면적(평)'] >= 25) &
    (df_apart['시군구'].str.contains("제주시 아라일동"))
].reset_index()
filtered_df = filtered_df.drop('index', axis=1)
display(filtered_df)
```

```python
# 텍스트 데이터 생성
def summarize_apart(df):
    summaries = []

    grouped = df.groupby('단지명')

    for name, group in grouped:
        시군구 = group.iloc[0]['시군구']
        거래건수 = len(group)
        평균평당가 = round(group['평당가격'].mean())

        # 최신 거래
        최신거래 = group.sort_values(by=['계약년도', '계약월', '계약일'], ascending=False).iloc[0]
        최신면적 = 최신거래['전용면적(평)']
        최신가격 = 최신거래['거래금액'] / 1e8
        최신년월 = f"{최신거래['계약년도']}년 {최신거래['계약월']}월"

        # 과거 거래 (6개월 이상 전 데이터 중 가장 최근)
        group['계약연월'] = group['계약년도'] * 100 + group['계약월']
        최신연월 = 최신거래['계약년월']
        과거거래 = group[group['계약년월'] <= 최신연월 - 6]  # 최소 6개월 전

        if not 과거거래.empty:
            기준거래 = 과거거래.sort_values(by='계약연월', ascending=False).iloc[0]
            과거가격 = 기준거래['거래금액']
            상승률 = ((최신거래['거래금액'] - 과거가격) / 과거가격) * 100
            if 상승률 > 0:
                상승내용 = f"6개월 전 대비 약 {상승률:.1f}% 상승했습니다."
            else:
                상승내용 = f"6개월 전 대비 약 {abs(상승률):.1f}% 하락했습니다."
        else:
            상승내용 = "최근 6개월간의 거래 내역이 없습니다."

        text = (
            f"{name}아파트는 {시군구}에 있으며, 최근 6개월간 {거래건수}건의 거래가 있었고 "
            f"평균 평당 가격은 {평균평당가:,}원입니다. "
            f"가장 최근 거래는 {최신년월}, {최신면적:.1f}평, {최신가격}억 원에 체결되었습니다. "
            f"{상승내용}"
        )
        print(text)

        summaries.append({
            "단지명": name,
            "요약문": text
        })

    return pd.DataFrame(summaries)


summary_df_apart = summarize_apart(df_apart)
summary_df_apart
```

```python
# 텍스트 데이터 생성
def summarize_office(df):
    summaries = []

    grouped = df.groupby('단지명')

    for name, group in grouped:
        시군구 = group.iloc[0]['시군구']
        거래건수 = len(group)
        평균평당가 = round(group['평당가격'].mean())

        # 최신 거래
        최신거래 = group.sort_values(by=['계약년도', '계약월', '계약일'], ascending=False).iloc[0]
        최신면적 = 최신거래['전용면적(평)']
        최신가격 = int(최신거래['거래금액'] / 1e4)  # 억 단위
        최신년월 = f"{최신거래['계약년도']}년 {최신거래['계약월']}월"

        # 과거 거래 (6개월 이상 전 데이터 중 가장 최근)
        group['계약연월'] = group['계약년도'] * 100 + group['계약월']
        최신연월 = 최신거래['계약년월']
        과거거래 = group[group['계약년월'] <= 최신연월 - 6]  # 최소 6개월 전

        if not 과거거래.empty:
            기준거래 = 과거거래.sort_values(by='계약연월', ascending=False).iloc[0]
            과거가격 = 기준거래['거래금액']
            상승률 = ((최신거래['거래금액'] - 과거가격) / 과거가격) * 100
            if 상승률 > 0:
                상승내용 = f"6개월 전 대비 약 {상승률:.1f}% 상승했습니다."
            else:
                상승내용 = f"6개월 전 대비 약 {abs(상승률):.1f}% 하락했습니다."
        else:
            상승내용 = "이전 거래와의 비교 정보는 부족합니다."

        text = (
            f"{name} 오피스텔은 {시군구}에 있으며, 최근 6개월간 {거래건수}건의 거래가 있었고 "
            f"평균 평당 가격은 {평균평당가:,}원입니다. "
            f"가장 최근 거래는 {최신년월}, {최신면적:.1f}평, {최신가격}억 원에 체결되었습니다. "
            f"{상승내용}"
        )

        print(text)
        
        summaries.append({
            "단지명": name,
            "요약문": text
        })

    return pd.DataFrame(summaries)


summary_df_office = summarize_office(df_office)
print(summary_df_office)
summary_df_office = summary_df_office[~summary_df_office['단지명'].str.contains('호텔')].reset_index()
summary_df_office = summary_df_office.drop('index',axis=1)
print(summary_df_office)
print(summary_df_office.shape)
```

```python
def build_documents(df, summary_df):
    documents = []

    # '계약년월일' 만들기 (정렬용)
    df['계약년월일'] = df['계약년도'].astype(str) + '-' + \
                      df['계약월'].astype(str).str.zfill(2) + '-' + \
                      df['계약일'].astype(str).str.zfill(2)

    grouped = df.groupby('단지명')

    for idx, row in summary_df.iterrows():
        complex_name = row['단지명']
        summary_text = row['요약문']

        if complex_name in grouped.groups:
            group = grouped.get_group(complex_name)
            group = group.sort_values(by=['계약년도', '계약월', '계약일'], ascending=False)

            # 거래내역 만들기
            history_lines = []
            for _, r in group.iterrows():
                date = f"{r['계약년도']}년 {r['계약월']}월 {r['계약일']}일"
                area = f"{r['전용면적(평)']:.1f}평"
                price = f"{r['거래금액'] / 1e8:.1f}억 원"
                line = f"- {date} | {area} | {price}"
                history_lines.append(line)

            history_text = "\n".join(history_lines[:20])  # 최근 20건까지만 출력 (원하면 늘릴 수 있음)

            full_doc = f"[요약]\n{summary_text}\n\n[거래내역]\n{history_text}"
            documents.append({
                "단지명": complex_name,
                "내용": full_doc
            })

    return pd.DataFrame(documents)

apart_final_df = build_documents(df_apart, summary_df_apart)
print(apart_final_df.head(5))
office_final_df = build_documents(df_office, summary_df_office)
print(office_final_df.head(5))
```

```python
# 스타벅스 리스트에 위도, 경도 넣기
import requests
import pandas as pd
import time

KAKAO_API_KEY = "-"

def get_lat_lng(address, api_key):
    url = "https://dapi.kakao.com/v2/local/search/address.json"
    headers = {"Authorization": f"KakaoAK " + api_key}
    params = {"query": address}
    response = requests.get(url, headers=headers, params=params)
    if response.status_code == 200:
        result = response.json()
        if result['documents']:
            x = result['documents'][0]['x']
            y = result['documents'][0]['y']
            return y, x
    return None, None

starbucks_df = pd.read_csv('starbucks_list.csv', encoding='utf-8')
# 스타벅스 주소 데이터프레임에 위경도 붙이기
starbucks_df["위도"] = None
starbucks_df["경도"] = None

for i in starbucks_df.index:
    address = starbucks_df.loc[i, "주소"]
    lat, lng = get_lat_lng(address, KAKAO_API_KEY)
    starbucks_df.at[i, "위도"] = lat
    starbucks_df.at[i, "경도"] = lng
    time.sleep(0.3)  # 과도한 요청 방지

starbucks_df
```

```python
import pandas as pd
import requests
import time

# 1. 주소 컬럼 만들기 (도로명 주소 우선, 없으면 번지로 대체 가능)
def make_address(row):
    if pd.notnull(row["도로명"]) and row["도로명"] != "-":
        return row["시군구"] + " " + row["도로명"]
    else:
        return row["시군구"] + " " + row["번지"]

df_apart["주소"] = df_apart.apply(make_address, axis=1)

# 2. 단지명 기준으로 그룹핑 → 대표 주소 추출
df_address = df_apart.groupby("단지명")["주소"].first().reset_index()

df_address
```

```python
# 3. Kakao Geocoding API로 위도/경도 가져오기
KAKAO_API_KEY = "-"

def get_lat_lng(address, api_key):
    url = "https://dapi.kakao.com/v2/local/search/address.json"
    headers = {"Authorization": f"KakaoAK {api_key}"}
    params = {"query": address}
    response = requests.get(url, headers=headers, params=params)
    if response.status_code == 200:
        result = response.json()
        if result["documents"]:
            x = result["documents"][0]["x"]
            y = result["documents"][0]["y"]
            return y, x
    return None, None
```

```python
# 4. 위경도 컬럼 추가
df_address["위도"] = None
df_address["경도"] = None

for i in df_address.index:
    addr = df_address.loc[i, "주소"]
    lat, lng = get_lat_lng(addr, KAKAO_API_KEY)
    df_address.at[i, "위도"] = lat
    df_address.at[i, "경도"] = lng
    time.sleep(0.3)  # 호출 제한 방지

# 또는 summary_df_apart에 병합해도 됨
summary_df_apart = apart_final_df.merge(df_address, on="단지명", how="left")

summary_df_apart
```
