### 1

제주도 부동산 실거래가 데이터(CSV)를 Pandas로 로드하고,
모델링에 적합하도록 데이터 정제(결측치, 타입 변환) 및 특성 공학(거래일로부터 연/월 추출, 면적 단위 통일 등)을 수행

```python
import pandas as pd
# 2016년 1월 ~ 2025년 7월 9일까지의 데이터
df0 = pd.read_csv('오피스텔_매매_실거래가_2016.csv',encoding="utf-8")
df1 = pd.read_csv('오피스텔_매매_실거래가_2017.csv',encoding="utf-8")
df2 = pd.read_csv('오피스텔_매매_실거래가_2018.csv',encoding="utf-8")
df3 = pd.read_csv('오피스텔_매매_실거래가_2019.csv',encoding="utf-8")
df4 = pd.read_csv('오피스텔_매매_실거래가_2020.csv',encoding="utf-8")
df5 = pd.read_csv('오피스텔_매매_실거래가_2021.csv',encoding="utf-8")
df6 = pd.read_csv('오피스텔_매매_실거래가_2022.csv',encoding="utf-8")
df7 = pd.read_csv('오피스텔_매매_실거래가_2023.csv',encoding="utf-8")
df8 = pd.read_csv('오피스텔_매매_실거래가_2024.csv',encoding="utf-8")
df9 = pd.read_csv('오피스텔_매매_실거래가_2025.csv',encoding="utf-8")
df = pd.concat([df0, df1, df2, df3, df4, df5, df6, df7, df8, df9]).reset_index()
# print(df.shape)

# 필요한 컬럼만 뽑기
df = df[['시군구', '전용면적(㎡)', '계약년월', '계약일' , '거래금액(만원)', '건축년도']]

# 거래금액의 쉼표(,) 제거 후 숫자 타입 변환
# print(df.dtypes)
df['거래금액'] = df['거래금액(만원)'].str.replace(',', '').astype('int')*10000
df = df.drop('거래금액(만원)', axis=1)

# 계약년월에서 '년', '월' 파생변수 생성
df['계약년도'] = df['계약년월']//100
df['계약월'] = df['계약년월']%100
df['계약년월'] = pd.to_datetime(df['계약년월'].astype(str), format='%Y%m') # 년-월 형태로 변경
df['계약년월'] = df['계약년월'].dt.to_period('M')
# print(df.head(5))

# '시'컬럼 생성(제주시/서귀포시)
df['시'] = df['시군구'].str.split(' ').str[1]
# 읍/면/동으로 그룹화해보기
df['읍면동'] = df['시군구'].str.split(' ').str[2]
df = df.drop('시군구', axis=1)

# 결측치확인
#print(df.isna().sum()) # 결측값 없음

display(df.head(10))
# 정제된 데이터 jeju_real_estate_cleaned.csv로 저장
df.to_csv('jeju_real_estate_cleaned.csv', index=False)
```

### 2
정제된 데이터를 사용하여 제주시/서귀포시 등 지역별, 전용면적 구간별 평당(또는 제곱미터당) 가격의 평균을 계산하고, 시간(연/월)에 따른 가격 변화 추이를 선 그래프 등으로 시각화하여 분석

```python
import matplotlib.pyplot as plt

# 한글 깨짐 방지(나눔고딕)
plt.rcParams['font.family'] = 'NanumGothic'

# 1. '평당가격' 파생 컬럼 생성
df = pd.read_csv('jeju_real_estate_cleaned.csv', encoding="utf-8")
df['평'] = df['전용면적(㎡)'] * 0.3025
df['평당가격'] = df['거래금액']//df['평']

# 2. 서귀포시/제주시 지역별 계약년월별 평균 '평당가격' 계산
new_df = df.groupby(['시', '계약년도'])['평당가격'].mean().reset_index()
print(new_df)

# 3. 전용면적을 특정구간별로 나누고 각 구간별 연도별 평균 '평당가격' 계산
# 65㎡미만, 65㎡이상~99㎡미만, 99㎡이상~135㎡미만, 135㎡이상~165㎡미만, 165㎡초과
# 전용면적을 구간별로 나누기
bins = [0, 65, 99, 135, 165, float('inf')]
labels = ['65㎡미만', '65㎡이상~99㎡미만', '99㎡이상~135㎡미만', '135㎡이상~165㎡미만', '165㎡초과']  # 구간 레이블
df['구간'] = pd.cut(df['전용면적(㎡)'], bins=bins, labels=labels, right=False)

# 연도별로 구간과 '평당가격' 평균 계산
grouped_df = df.groupby(['구간', '계약년도'])['평당가격'].mean().reset_index()
print(grouped_df)
grouped_df.isna().sum() # 구간으로 나눴을 때 평당가격이 없는 case 존재 -> 거래가 없었음. (평수가 넓은 부동산일수록 거래량이 적거나 없기도 함)
```

```python
# 4. 2와 3의 결과를 선 그래프로 시각화하여 가격 추이 확인
#2번 결과: 지역별 연도별 평균 '평당가격' 시각화
plt.figure(figsize=(20, 6))
for city in new_df['시'].unique():
    city_data = new_df[new_df['시'] == city]
    plt.plot(city_data['계약년도'], city_data['평당가격'], label=f'{city} 평당가격', marker='o')

plt.title('지역별 연도별 평균 평당가격 추이')
plt.xlabel('연월')
plt.ylabel('평당가격')
plt.ticklabel_format(style='plain', axis='y')  # 지수표기법 비활성화
plt.legend(loc='best', bbox_to_anchor=(1.0,1.0))
plt.show()
```

```python
# 4. 2와 3의 결과를 선 그래프로 시각화하여 가격 추이 확인
#2번 결과: 읍면동별 연도별 평균 '평당가격' 시각화
new_df = df.groupby(['시', '읍면동', '계약년도'])['평당가격'].mean().reset_index()
plt.figure(figsize=(55, 10))  # 가로폭 넓게 설정 (많은 읍면동 대비)

# 읍면동 기준 반복
for dong in new_df['읍면동'].unique():
    dong_data = new_df[new_df['읍면동'] == dong]
    plt.plot(dong_data['계약년도'], dong_data['평당가격'], label=f'{dong}', marker='o')

plt.title('읍면동별 계약년월별 평균 평당가격 추이')
plt.xlabel('계약년도')
plt.ylabel('평당가격')
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=45)
plt.legend(title='읍면동', bbox_to_anchor=(1.0, 1.0), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# 읍면동별 거래량 히스토그램
import matplotlib.pyplot as plt
import seaborn as sns

# 읍면동별 거래량 집계
dong_count = df['읍면동'].value_counts().reset_index()
dong_count.columns = ['읍면동', '거래량']

# 히스토그램 시각화
plt.figure(figsize=(14, max(6, len(dong_count) * 0.3)))  # 읍면동 수에 따라 높이 자동 조절
sns.barplot(data=dong_count.sort_values('거래량', ascending=True),
            x='거래량', y='읍면동', hue='읍면동', palette='viridis')

plt.title('읍면동별 오피스텔 거래량 히스토그램')
plt.xlabel('거래 건수')
plt.ylabel('읍면동')
plt.tight_layout()
plt.show()
```

```python
# 평당 가격 높은 지역 상위 5개 읍면동
# 읍면동별로 평당가격의 max - min 계산
import matplotlib.pyplot as plt

dong_range = df.groupby('읍면동')['평당가격'].agg(lambda x: x.max() - x.min())
top_dongs = dong_range.sort_values(ascending=False).head(5).index

filtered_df = df[df['읍면동'].isin(top_dongs)]
dong_df = filtered_df.groupby(['읍면동', '계약년도'])['평당가격'].mean().reset_index()

plt.figure(figsize=(55, 8))
for dong in dong_df['읍면동'].unique():
    dong_data = dong_df[dong_df['읍면동'] == dong]
    plt.plot(dong_data['계약년도'], dong_data['평당가격'], label=dong, marker='o')

plt.title('평당가격 변동폭 상위 5개 읍면동 추이 (max - min)')
plt.xlabel('계약년도')
plt.ylabel('평당가격')
plt.xticks(rotation=45)
plt.ticklabel_format(style='plain', axis='y')
plt.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
plt.tight_layout()
plt.show()
```

```python
# 전용면적 구간별 연도별 평균 '평당가격'
plt.figure(figsize=(55, 10))
for label in labels:
    segment_data = grouped_df[grouped_df['구간'] == label]
    plt.plot(segment_data['계약년도'], segment_data['평당가격'], label=f'{label} 평당가격', marker='x')

plt.title('전용면적 구간별 연도별 평균 평당가격 추이')
plt.xlabel('연도')
plt.ylabel('평당가격')
plt.ticklabel_format(style='plain', axis='y')  # 지수표기법 비활성화
plt.legend(loc='best')
plt.show()
```

### 3

전처리된 데이터 바탕으로 특정 부동산의 가격 예측하는 회귀모델(선형회귀, 랜덤포레스트 등)을 사이킷런을 통해 구축하고, 모델의 성능을 평가해 어떤 특성이 가격 예측에 중요한 영향을 미치는지 확인

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# 거래금액을 타겟(y)으로, 나머지(지역, 면적, 계약년월)를 특성(x)으로 설정.
df = pd.read_csv('jeju_real_estate_cleaned.csv', encoding='utf-8')
target = df['거래금액']
x = df.loc[:, df.columns != '거래금액']
x = x.drop(['계약년월', '계약일'], axis=1)
x_town = x[['시']]
x_city = x[['읍면동']]

# 지역 등 범주형 특성 원 핫 인코딩
# display(x.dtypes)
# onehot = OneHotEncoder(sparse=False)

# ohe_encoded = onehot.fit_transform(x_town)
# ohe_df = pd.DataFrame(ohe_encoded, columns=onehot.get_feature_names_out(['시'])) # 원 핫 인코딩된 결과를 DataFrame으로 변환
# # 원 핫 인코딩 결과를 기존 특성과 결합
# x = x.drop('시', axis=1)  # '시' 컬럼은 제거
# x = pd.concat([x, ohe_df], axis=1)

# onehot = OneHotEncoder(sparse=False)

# ohe_encoded = onehot.fit_transform(x_city)
# ohe_df = pd.DataFrame(ohe_encoded, columns=onehot.get_feature_names_out(['읍면동'])) # 원 핫 인코딩된 결과를 DataFrame으로 변환
# # 원 핫 인코딩 결과를 기존 특성과 결합
# x = x.drop('읍면동', axis=1)  # '읍면동' 컬럼은 제거
# x = pd.concat([x, ohe_df], axis=1)


# 지역 등 범주형 특성 라벨 인코딩

# 시 인코딩
le = LabelEncoder()
x['시'] = le.fit_transform(x['시'])
x['시'].head(10)
le = LabelEncoder()
x['읍면동'] = le.fit_transform(x['읍면동'])
x['읍면동'].head(10)

print("원핫인코딩, RMSE: 25316482.013933763 R²: 0.9537320047681924")
print("라벨인코딩, RMSE: 29212885.847279545 R²: 0.9383940167028179")

# 데이터를 학습/테스트용으로 분리
X_train, X_test, y_train, y_test = train_test_split(x, 
                                                    target, 
                                                    test_size=0.2, 
                                                    random_state=42)

# RandomForestRegressor 모델 학습 및 예측
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test) # 예측

# 테스트 데이터에 대한 RMSE 및 R² 점수 계산 및 출력
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
print(f"RMSE: {rmse}")
print(f"R²: {r2}")

# 학습된 모델의 특성 중요도 상위 5개 출력.
feature_importances = model.feature_importances_
indices = np.argsort(feature_importances)[::-1]  # 중요도에 따라 내림차순 정렬
print("\n상위 5개 특성 중요도:")
for f in range(5):
    print(f"{f + 1}. {x.columns[indices[f]]}: {feature_importances[indices[f]]}")
```
